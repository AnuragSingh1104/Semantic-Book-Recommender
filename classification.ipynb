```python
books.drop(['subtitle','missing_description','ageofbook','words_in_description'],axis=1,inplace=True)
```

```python
temp_df=books["categories"].value_counts().reset_index().query("count>50")
```

```python
category_mapping={"Fiction":"Fiction",
                  "Juvenile Fiction":"Children's Fiction",
                  "History":"NonFiction",
                  "Biography & Autobiography":"NonFiction",
                  "Literary Criticism":"NonFiction",
                  "Religion":"NonFiction",
                  "Philosophy":"NonFiction",
                  "Comics & Graphic Novels":"Fiction",
                  "Science":"NonFiction",
                  "Poetry":"Fiction",
                  "Drama":"Fiction",
                  "Juvenile Nonfiction":"Children's NonFiction",
                  
                  }

books["simple_categories"]=books["categories"].map(category_mapping)
```

```python
from transformers import pipeline

fiction_categories = ['Fiction', "NonFiction"]

pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
sequence = books.loc[books["simple_categories"]=="Fiction","description"].reset_index(drop=True)[0]
```

```python
pipe(sequence,fiction_categories)
```

```python
import numpy as np
max_index=np.argmax(pipe(sequence,fiction_categories)["scores"])
max_label=pipe(sequence,fiction_categories)["labels"][max_index]
max_label
```

```python
def generate_prediction(sequence, categories):
    predictions= pipe(sequence, categories)
    max_index=np.argmax(predictions["scores"])
    max_label=predictions["labels"][max_index]
    return max_label
```

## checking accuracy of classification

```python
from tqdm import tqdm

actual_categories = []
predicted_categories = []

# Filter data first
fiction_descriptions = books.loc[books["simple_categories"]=="Fiction", "description"]
for i in tqdm(range(0,300)):
    sequence = fiction_descriptions.reset_index(drop=True)[i]  # Safe positional access
    predicted_categories.append(generate_prediction(sequence, fiction_categories))
    actual_categories.append("Fiction")
```

```python
fiction_descriptions = books.loc[books["simple_categories"]=="NonFiction", "description"]
for i in tqdm(range(0,300)):
    sequence = fiction_descriptions.reset_index(drop=True)[i]  # Use .iloc for positional access
    predicted_categories.append(generate_prediction(sequence, fiction_categories))
    actual_categories.append("NonFiction")
```

```python
prediction_df=pd.DataFrame({"actual":actual_categories,"predicted":predicted_categories})
```

```python
prediction_df.head()
```

```python
prediction_df["Correct_prediction"]=np.where(prediction_df["actual"]==prediction_df["predicted"],1,0)
```

```python
isbn=[]
predicted_categories=[]

missing_cats=books.loc[books["simple_categories"].isna(),["isbn13","description"]].reset_index(drop=True) 
```

```python
for i in tqdm(range(0,len(missing_cats))):
    sequence = missing_cats["description"].reset_index(drop=True)[i]
    predicted_categories.append(generate_prediction(sequence, fiction_categories))
    isbn.append(missing_cats["isbn13"].reset_index(drop=True)[i])
```

```python
missing_predicted_df=pd.DataFrame({"isbn13":isbn,"predicted_category":predicted_categories})
```
