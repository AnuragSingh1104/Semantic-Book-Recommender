```python
import numpy as np 
books["missing_description"]=np.where(books["description"].isna(),1,0)
books["ageofbook"]=2025-books['published_year']
```

```python
columns_of_interest=['num_pages','ageofbook','missing_description','average_rating']
```

```python
correlation_matrix=books[columns_of_interest].corr(method='spearman')
```

```python
sns.set_theme(style='white')
plt.figure(figsize=(8,6))
heatmap=sns.heatmap(correlation_matrix,annot=True,fmt="0.2f")
```

- No columns have high correlation 

## Removing rows which have null value in column in intersts

```python
books[(books['description'].isna())|
                            (books['average_rating'].isna())|
                            (books['published_year'].isna())|
                            (books['num_pages'].isna())]
```

- number of missing rows is less we can drop them

```python
books_after_removing=books.dropna(subset=['num_pages','description','published_year','average_rating'])
```

```python
books_after_removing
```

```python
books_after_removing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
books_after_removing['words_in_description']=books_after_removing["description"].str.split().str.len()
```

```python
books_after_removing
```

```python
sns.histplot(books_after_removing['words_in_description'],bins=30,kde=True)
```

```python
books_after_removing[(books_after_removing['words_in_description']>4 )&(books_after_removing['words_in_description']<15)] 
```

```python
books_after_removing[(books_after_removing['words_in_description']>15 )&(books_after_removing['words_in_description']<24)] 
```

```python
books_after_removing[(books_after_removing['words_in_description']>25 )&(books_after_removing['words_in_description']<34)] 
```

- 25 and above will be the cutoff in description

```python
books_after_description=books_after_removing[books_after_removing['words_in_description']>=25]
```

```python
books_after_description
```

```python
books_after_description["title_and_subtitle"]=(
    np.where(books_after_description['subtitle'].isna(),books_after_description['title'],
            books_after_description[["title","subtitle"]].astype(str).agg(":".join,axis=1),)
)
```

```python
books_after_description
```

```python
# Correct method using lambda
books_after_description["tagged_description"] = books_after_description[['isbn13','description']].astype(str).agg(lambda x: ' '.join(x), axis=1)
```

```python
books_after_description
```

- Removing unwanted columns

```python
books_after_description.drop(['subtitle','missing_description','ageofbook','words_in_description'],axis=1)
```

```python
books_after_description.to_csv('Books_cleaned.csv',index=False)
```
